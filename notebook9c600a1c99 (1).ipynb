```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://urldefense.com/v3/__https://github.com/kaggle/docker-python*5Cn__;JQ!!DZ3fjg!rkUfiF95cCb-WQ59VBkjFQWtcVthDwqpI-ALjht_QCjtwqzmUI-oOoV0sqNkFhzUKPw$ # For example, here's several helpful packages to load


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import statistics


# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os

# For question 3, we assumed the average size of an image to be 7395 KB * 1024 bytes per MB.
# This value was Googled based on Twitter data regarding the sizes of videos in tweets.
AVG_VIDEO_SIZE = 7395 * 1024

# For question 3, we assumed the average size of an image to be 141 KB * 1024 bytes per MB.
# This value was Googled based on Twitter data regarding the sizes of images in tweets.
AVG_PIC_SIZE = 141 * 1024

avg_likes = []
avg_view = []
avg_byte = []
avg_viewed = []

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        df = pd.read_csv(os.path.join(dirname, filename), index_col=0)
#### QUESTION 1 ####
# The average number of people that like an Elon Musk tweet was computed in the following way:
# For each CSV file, each entry's "nlikes" value was appended to a list. Once all values were
# appended across all CSV files (over all years), the mean of the list was taken and the value
# was rounded.

        avg_likes = avg_likes + df['nlikes'].tolist()
        
#### QUESTION 2 ####
# The average number of people that view one Elon Musk tweet was computed in the following way:
# For each tweet (for each row in each CSV file), the "nlikes", "nreplies", and "nretweet"
# values were added up and dividing by 0.05. This value would be the number of views for
# a single tweet. This calculation was done for each tweet, the values were all appended to
# a list, and the mean of all values in the list was computed. This numerical value was then
# rounded.
        df['views']= df.loc[:, ['nlikes','nreplies', 'nretweets']].sum(axis = 1)/.05
        avg_view += df['views'].tolist()
        
#### QUESTION 3 ####
# The average amount of data in MB for each Elon Musk tweet was computed in the following way:
# After analyzing the number of bytes in a tweet with a video/picture vs the tweets without a video/picture,
# we assumed that the tweets with videos/pictures would heavily sway the average up. For this reason,
# since we are merely trying to find an estimate, we only considered the tweets with videos/pictures.
# For each of these tweets, we took the AVG_PIC_SIZE and AVG_VIDEO_SIZE constants and added it
# to a list. Once an estimate of the number of bytes for each tweet is added to a list, the
# mean of the list is computed and the answer is rounded.
        avg_byte += df['thumbnail'].tolist()
        
#### QUESTION 5 ####
# The number of Elon Musk tweets that are viewed per minute was computed in the following way:
# Per CSV file (per year), the total number of tweets was obtained, and then divided by 365.25
# (to get the daily number of tweets), then divided by 24 (to get the hourly number of tweets), then
# divided by 60 (to get the numbe of tweets per minute). This value was then appended to a list and
# the mean of the list was computed, before rounding the answer.
        l = ((len(df.index) / 365.25) / 24) / 60
        avg_viewed = avg_viewed + [l]
    

q1 = round(statistics.mean(avg_likes))
print(q1)
q2 = round(statistics.mean(avg_view))
print(q2)

q3sum = 0
for i in avg_byte:
    if i != i:
        continue
    if 'video' in i:
        q3sum += AVG_VIDEO_SIZE
    else:
        q3sum += AVG_PIC_SIZE
q3sum = round(q3sum/len(avg_byte))
print(q3sum)

#### QUESTION 4 ####
# The total data transfer involved in displaying an Elon Musk tweet was computed by
# taking the result of Q3 (the average amount of data stored per Elon Musk tweet) and
# multiplying it by the result of Q2 (the average number of people that view an Elon
# Musk tweet).
q4 = q3sum * q2
print(q4)

for j in range(0, len(avg_viewed)):
    avg_viewed[j] = avg_viewed[j] * q3sum

q5 = round(statistics.mean(avg_viewed))
print(q5)

#### QUESTION 6 ####
# The amount of data accessed per minute in the process of displaying Elon Musk tweets was
# computed in the following way: The result of Q4 (the total data transfer involved in displaying
# an Elon Musk tweet) was divided by the result of Q5 (the number of tweets that are viewed per
# minute).
q6 = round(q4/q5)
print(q6)


#Q7
#If the likes in a tweet is modified, then it is safe to assume that everybody viewing the tweet would see it modified. 
#People who are viewing the tweet at the current time will see the likes number updated, affecting their data. As such,
#we assume that question 4 is a good indicator of the data touched, since it needs to be constantly transferred. We estimated 
#there is 17739191536 bytes transferred per tweet. Using a random sample of months, we found that Musk tweets approx 20 times a month.
#Most engagement would be done within the month, so 17739191536 * 20 would give us the data transferred. This value
#is 354783830720. 

        

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
