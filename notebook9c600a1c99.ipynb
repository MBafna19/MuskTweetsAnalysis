```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import statistics


# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os

AVG_VIDEO_SIZE = 7395 * 1024
AVG_PIC_SIZE = 141 * 1024



avg_likes = []
avg_view = []
avg_byte = []
avg_viewed = []

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        df = pd.read_csv(os.path.join(dirname, filename), index_col=0)
        #q1
        avg_likes = avg_likes + df['nlikes'].tolist()
        
        #q2
        df['views']= df.loc[:, ['nlikes','nreplies', 'nretweets']].sum(axis = 1)/.05
        avg_view += df['views'].tolist()
        
        #q3
        avg_byte += df['thumbnail'].tolist()
        
        #q5
        l = ((len(df.index) / 365.25) / 24) / 60
        avg_viewed = avg_viewed + [l]
    

q1 = round(statistics.mean(avg_likes))
print(q1)
q2 = round(statistics.mean(avg_view))
print(q2)

q3sum = 0
for i in avg_byte:
    if i != i:
        continue
    if 'video' in i:
        q3sum += AVG_VIDEO_SIZE
    else:
        q3sum += AVG_PIC_SIZE
q3sum = round(q3sum/len(avg_byte))
print(q3sum)
q4 = q3sum * q2
print(q4)

for j in range(0, len(avg_viewed)):
    avg_viewed[j] = avg_viewed[j] * q3sum
q5 = round(statistics.mean(avg_viewed))
print(q5)

q6 = round(q4/q5)
print(q6)


#Q7
#If the likes in a tweet is modified, then it is safe to assume that everybody viewing the tweet would see it modified. 
#People who are viewing the tweet at the current time will see the likes number updated, affecting their data. As such,
#we assume that question 4 is a good indicator of the data touched, since it needs to be constantly transferred. We estimated 
#there is 17739191536 bytes transferred per tweet. Using a random sample of months, we found that Musk tweets approx 20 times a month.
#Most engagement would be done within the month, so 17739191536 * 20 would give us the data transferred. This value
#is 354783830720. 

        

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
